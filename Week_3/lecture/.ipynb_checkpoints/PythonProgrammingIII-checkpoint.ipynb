## Types of Bugs

Bugs fall into two (very broad) categories:

- **General Errors**: The code does not run - it stops and gives you some kind of error message.
    - The exact message depends on what went wrong. These are sometimes specific to modules.
    - These are further divided into error types (i.e. Syntax Errors).

- **Logical Errors**: The code does run - but it doesn't do what you expected it to do.
    - You intended to do something, and translated that intention into code instructions - but the instructions you wrote didn't match your intention.
    - No 'types' here - logical errors are specific to the exact problem you are trying to code. We will discuss some general methods to try to isolate where things are going wrong.

### Error Types + Examples

When we hit an error that stops our code from running, we will recieive a printout showing where the error occured, a message, and an **Error Type**. The error type is a broad category that gives us a hint as to what went wrong.

When we hit a logical error, the code will run, so we won't get a helpful message. Its up to us to figure out what went wrong.

There are many error types in Python, some more obscure than others. We'll cover some basic ones here.
- You don't have to memorize these, you'll become familiar with them as you program more!

## Runtime Errors

Runtime Errors are errors that are not detected until the code is run (hence the name), but the code execution encounters some condition that cannot be executed and requires a stop.

Runtime Errors are an umbrella term, with many different specific types.
- You will rarely encounter a specific "Runtime Error" message as this is only raised when an error does not fall under any other category
- However, any error that is detected when code is run can be considered a Runtime Error.

## Basic Debugging Strategies

So we have buggy code, what do we do?

There are many methods for debugging, everyone has their own preferences.
- "Rubber Ducky" Programming: Explaining exactly what each line of code does and what each variable holds to a Rubber Ducky on your desk until you find the logical issue
- Print Statements
- Assert Statements
- Advanced Tools (i.e. Python Debugger(pdb), IDE based debuggers)

We're going to focus mostly on the basic methods of **Print** and **Assert** today because they're simpler and more beginner friendly.

However, despite their simplicity, they're incredibly powerful. While some programmers swear by advanced tools, there are just as many others who find simple Print and Assert statements sufficient to write performative, complex code.

This is not to say debugger tools aren't useful, especially if the issue you're encountering is particularly arcane. ### Print

Print debugging involves inserting temporary `print` statements into a section of code, creating breakpoints that let us check the status of the code.

There are many different ways to use print statements depending on both the programmer and the problem at hand.

A few common examples:

- **Control Structures**: Print when code enters an if-else statement (or note if it doesn't)

```python
if x:
    print("Code got here")
```

- **Loop Execution**: Print each time a loop executes

```python
x = 0
for i in range(10):
    x += i
    print(f"Value of x: {x} after loop {i+1})")
```

- **Checking Values/ Type**: When we create an intermediate variable, we may want to check its value and type is as expected.

```python
print(f"Variable is of type {type(x)}, storing {x})")
```

- **Others**: error handling, timing code, checking conditions, etc.


> These are very general examples. In reality, you'll print **whatever you think you need to check to solve your problem**

You can choose to remove print statements once you've found and fixed a bug. Alternatively, you can wrap them in a conditional argument (i.e. `debug`), and leave them in place to save time if the function ever breaks again.
- This structure is yet another reason to functionalize your code

````python
name = 'Jeremy'

f'My name is {name}!'
````

````python
def calculate_average(numbers, DEBUG=False):
  """A function to average numbers with built in debugging"""
  total = 0
  count = 0
  for number in numbers:
    if DEBUG:
      print("Number:", number)
    total += number
    if DEBUG:
      print("Running Total:", total)
    count += 1
    if DEBUG:
      print("Count:", count)
  average = total / count
  return average
````

````python
my_list = [10, 20, 30, 40, 50]
result = calculate_average(my_list) # Normal Behaviour
# result = calculate_average(my_list, DEBUG=True) # Debugging Behaviour
print("Average:", result)
````

### Assert

Used to check to **make sure a function is behaving as expected**, or conditions are met.

Closely linked to the idea of unit testing/ test driven development
- Important concept in general software engineering.

If we know the expected inputs and outputs of the function we are trying to create, it can be extremely helpful.
- It is particularly useful for checking if edge case conditions are met
- Can be used outside of function to check input/ output
- Can be used inside of function to check for certain conditions/ raise errors


#### Asserting Internal Condition

````python
# This function assumes that scores is a list with at least 1 element
def calculate_average(scores):
  assert type(scores) == list, "Input must be a list" # condition, message if False
  assert len(scores) > 0, "List cannot be empty" # condition, message if False
  return sum(scores) / len(scores)
````

````python
print("Average:", calculate_average([95, 85, 90, 100])) # list
````

````python
print("Average:", calculate_average([])) # empty list
````

````python
print("Average:", calculate_average(10)) # One number
````

#### Asserting On Input/ Output

````python
def factorial(n):
  """Calculate the factorial of a non-negative integer n."""
  if n == 0:
    return 1
  elif n < 0:
    raise ValueError("Factorial is not defined for negative numbers.")
  result = 1
  for i in range(1, n + 1):
    result *= i
  return result
````

````python
factorial(-1)
````

**BEFORE** writing the actual function internals, write some checks to make sure the function is doing as expected
- Ensure to include normal behaviour and edge cases
- Edge cases here might be 0 (defined as 1), 1 (does it work for only 1 number?), -1 (should throw an error)
- Normal behaviour just try some whole number, i.e. $5! = 120$

````python
assert factorial(0) == 1, "Factorial of 0 should be 1"
assert factorial(3) == 6, "Factorial of 3 should be 6"
assert factorial(5) == 120, "Factorial of 5 should be 120"

# # More complex, we WANT an error:
try:
  factorial(-1) # If this raises an error go to `except`
  assert False, "Factorial of -1 did not raise an error" # If it didn't error in the function, print this
except ValueError:
  pass #If factorial(-1) throws an error thats what we want. Pass to prevent error from stopping code
````

Again, while this is helpful debugging, it can also become the basis for unit tests, to ensure this code always behaves as expected.
- Tests are run when code is changed to ensure incremental changes to a complex code base don't break anything
- They're usually in `.py` files. They can be called automatically when code is changed (thats a DevOps thing, not covered here)

````python
ages = [20, 25, 30, 35, 40, 45, 50]
incomes = [0, 50000, 75000, 90000, 1600000, 265000, 0]

def process_data(ages, incomes):
    """
    Calculate some summary statistics for lists of a person's incomes at different ages
    """
    # Get the average age
    total_age = 0
    for i in range(len(ages)):
        total_age += ages[i]
    count = range(len(ages))
    average_age = total_age / "count"

    # Get the median income
    sorted_incomes = incomes.sort() 
    mid_index = len(sorted_incomes) // 2 
    median_income = (sorted_incomes[mid_index + 1] + sorted_incomes[mid_index]) / 2

    # Get the total % change in income
    initial_income = incomes[0]
    final_income = incomes[-1]
    income_increase_percentage = 100 * (final_income / initial_income - 1)
    if initial_income == 0:
        income_increase_percentage = "N/A"
    income_increase_decimal = income_increase_percentage / 100 #convert back to decimal

    # Get the average income
    average_income = sum(incomes) / len(incomes) - 1

    # Final Income
    last_income = incomes[len(incomes)]

    return average_age, median_income, income_increase_percentage, income_increase_decimal, average_income, last_income
````

````python
try:
  results = process_data(ages, incomes)
  print("Average Age:", results[0])
  print("Median Income:", results[1])
  print("Income Increase Percentage:", results[2])
  print("Income Increase As Decimal:", results[3])
  print("Average Income:", results[4])
  print("Final Income:", results[5])
  assert results[0] == 35, "Average Age is Incorrect"
  assert results[1] == 75000, "Median Income is Incorrect"
  assert results[4] == sum(incomes)/len(incomes), "Average Income is Incorrect"
  assert results[5] == incomes[-1], "Final Income is Incorrect"
except Exception as e:
  print("Error occurred:", e)
````
